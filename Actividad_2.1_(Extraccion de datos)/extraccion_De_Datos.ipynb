```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
#carga de archivos .xlsx
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
df.info()
```

FILTROS BASICOS PARA VARIABLES 

```python
#FILTRO POR OBJETOS
#filtracion de aquellos productos que son de la categoria de "Nombre de cliente"
#y que su nombre de cliente sea "PÚBLICO EN GENERAL"
filtro1=df[(df["NOMBRE_CLIENTE"]=="PÚBLICO EN GENERAL")]
filtro1 
```

```python
filtro1.to_csv('filtro1.csv')
```

FILTRO POR FILAS

```python
#este filtro imprimira todas las filas consecutivamente desde la 8 hasta la 20
filtro2C=df.iloc[8:20:]    
filtro2C
```

```python
filtro2C.to_csv('filtro2C.csv')
```

```python
#Este filtro imprimira las filas 15 y 17
filtro2=df.iloc[[15,17],:]  
filtro2
```

```python
filtro2.to_csv('filtro2.csv')
```

FILTRO POR COLUMNAS

```python
#filtro por columnas CONSECUTIVAS, imprimira las columnas de la 4 a la 11
filtro3c=df.iloc[:,4:11]        
filtro3c
```

```python
filtro3c.to_csv('filtro3c.csv')
```

```python
#filtro por columnas no consecutivas, imprimira las columnas 3,6,9
filtro3=df.iloc[:,[3,6,9]]    #Columnas no consecutivas
filtro3
```

```python
filtro3.to_csv('filtro3.csv')
```

```python
df1=pd.read_excel('Detalle precios y productos fabricados 2022.xlsx',index_col=2) 
df1  
```

FILAS VS COLUMNAS

```python
#filtro que selecciona aquellas filas que tiene por indice 
#DIEGO ARMANDO LEON DURAN, LETICIA RAMIREZ HERNANDEZ
#  y que solo muestre la columna COSTO_TOTAL_CALCULADO
filtro4=df1.loc[["DIEGO ARMANDO LEON DURAN","LETICIA RAMIREZ HERNANDEZ"],["COSTO_TOTAL_CALCULADO"]]
filtro4
```

```python
filtro4.to_csv('filtro4.csv')
```

```python
filtro5.to_csv('filtro5.csv')
```

```python
#filtro por cola
#imprimira las ultimas 5 filas
filtro6=df1.tail()
filtro6
```

```python
filtro6.to_csv('filtro6.csv')
```

FILTROS BASICOS PARA VARIABLES CUANTITATIVAS

```python
#   filtro por condicion
#   imprimira aquellas filas que tengan un COSTO_UNITARIO mayor a 70000
filtro7=df[df["COSTO_UNITARIO"]>5000]
filtro7
```

```python
filtro7.to_csv('filtro7.csv')
```

FILTROS BASICOS CON CONECTORES LOGICOS

```python
#filtro &
#imprimira aquellas filas que tengan un MARGEN_UNITARIO_CALCULADO menor a 10000
#y que su fecha sea 2022-01-05
filtro8=df[(df["MARGEN_UNITARIO_CALCULADO"]<10000) & (df["FECHA_DOC"]=="2022-01-05")]
filtro8
```

```python
filtro8.to_csv('filtro8.csv')
```

```python
#filtro or
#imprimira aquellas filas que tengan un PRECIO_UNITARIO menor a 100
#o que su fecha sea 2022-12-23
filtro9=df[(df["PRECIO_UNITARIO"]<100) | (df["FECHA_DOC"]=="2022-12-23")]
filtro9
```

```python
filtro9.to_csv('filtro9.csv')
```

```python
#filtro not 
#imprimira aquellas filas que no tengan como nombre de cliente "PÚBLICO EN GENERAL"

filtro10=df[~(df["NOMBRE_CLIENTE"]=="PÚBLICO EN GENERAL")]
filtro10
```

```python
filtro10.to_csv('filtro10.csv')
```
