```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns   
```

```python
#carga de archivos .xlsx
df = pd.read_excel('Datos de Facturación.xlsx')
df
```

```python
df.info()
```

```python
#conviertimos los valores en la columna "CVE_CLPV" a un tipo de dato numérico (enteros o flotantes)
#y transformamos los valores que no se puedan convertir a NaN
df['CVE_CLPV'] = pd.to_numeric(df['CVE_CLPV'], errors='coerce')
#mostramos los valores de la columna "CVE_CLPV" que se encuentran entre 1000 y 2000
filtro1=filtro1 = df[(df["CVE_CLPV"]>=1000) & (df["CVE_CLPV"]<=2000)]
filtro1
```

```python
filtro1.to_csv('filtro1.csv')
```

```python
#mostramos todos los valores de la columna "CVE_VEND" que no sean 4 o 5
filtro2 = df[~((df["CVE_VEND"] == 5) | (df["CVE_VEND"] == 4))]
filtro2
```

```python
filtro2.to_csv('filtro2.csv')
```

```python
#convertimos los valores de la columna "FECHA_ENT" a un tipo de dato fecha, y especificamos el formato
#cualquier valor que no se pueda convertir a datetime según el formato especificado se transformará en NaN
df["FECHA_ENT"] = pd.to_datetime(df["FECHA_ENT"], errors="coerce", format='%d-%m-%Y %H:%M')
#mostramos todos los valores de la columna "FECHA_ENT" que sean iguales a "2022-02-28"
filtro3 = df[(df["FECHA_ENT"] == "2022-02-28")]
filtro3
```

```python
filtro3.to_csv('filtro3.csv')
```

```python
#mostramos todos los valores de la columna "CAN_TOT" que sean menores a 5951.7 o que tengan un "STATUS" igual a "E"
filtro4=df[(df["CAN_TOT"]<5951.7) | (df["STATUS"]=="E")]
filtro4
```

```python
filtro4.to_csv('filtro4.csv')
```

```python
#mostramos solo las columnas "CVE_DOC", "FECHA_ENT", "FECHA_VEN" y "CAN_TOT"
filtro5=df.loc[:,["CVE_DOC", "FECHA_ENT", "FECHA_VEN" , "CAN_TOT"]]    #Columnas no consecutivas
filtro5
```

```python
filtro5.to_csv('filtro5.csv')
```

```python
#mostramos solo las filas 7001 a 7099
#colocamos 7100 para que se incluya la fila 7099
filtro6=df.iloc[7001:7100:]
filtro6
```

```python
filtro6.to_csv('filtro6.csv')
```
