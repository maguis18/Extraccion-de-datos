```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#leemos el archivo csv
df = pd.read_csv('Bozeman_United States.csv')
df
```

```python
df.info()
#damos una vista a los datos
```

USO DE OR (CONECTORES LOGICOS)

```python
#los registros de 5 "host name" diferentes 
#para esto solo necesitamos la columna "host_name"
#filtro1=df.head() pero como existe la posibilidad de que las primeras 5 lineas tengan al mismo host , mejor usamos 
filtro1=df[(df['host_name'] == 'Katie') | (df['host_name'] == 'Irene') | (df['host_name'] == 'Steven And Eleanor') | (df['host_name'] == 'Allison') | (df['host_name'] == 'Natalie')]
filtro1
```

Uso del operador logico para comparar

```python
#df["host_since"] = pd.to_datetime(df['host_since'], errors='coerce')
#los registros de los host que se hayan registrado antes del 2019
#para esto solo necesitamos la columna "host_since" y la fecha a comparar
filtro2=df[df["host_since"] < "2019-01-01"]
filtro2
```

Uso de or (CONECTORES LOGICOS)

```python
#los registros de los host que tarden en responder menos de un día
#para esto solo necesitamos la columna "host_response_time" y el valor a comparar
#que en este caso es within a day
filtro3 = df[(df['host_response_time'] == 'within a day') | (df['host_response_time'] == 'within an hour') | (df['host_response_time'] == 'within a few hours')]
filtro3
```

Uso del operador logico para comparar

```python
#los registros que tengan al menos 5 acomodaciones
#para esto solo necesitamos la columna "accommodates" y el valor a comparar
filtro4=df[df["accommodates"]>=5]
filtro4
```

Uso de CONECTORES LOGICOS

```python
#los registros de los tipos de cuarto "Entire home/apt" y "Private room"
#para esto solo necesitamos la columna "room_type" y los valores que queremos que se impriman 
#usamor or ya que queremos los registros que tengan uno de los dos valores
filtro5=df[(df["room_type"]=="Entire home/apt") | (df["room_type"]=="Private room")]
filtro5
```

uso de operadores logicos

```python
#los registros que tengan a lo mucho 4 camas
#para esto solo necesitamos la columna "bedrooms" e indicar la operacion 
filtro6=df[df["bedrooms"]<=4]
filtro6
```

uso de CONECTORES LOGICOS or &

```python
#Los registros que tengan 1 bathroom o que sean superhost pero con la condicion pero con la condicion 
# de que todos sean instant_bookable
#para esto solo necesitamos las columnas "bathrooms", "host_is_superhost" y "instant_bookable"
#usamos el operador & para que se cumplan las dos condiciones
filtro7=df[(df["bathrooms"]==1) | (df["host_is_superhost"]=="t") & (df["instant_bookable"]=="t")]
filtro7
```

uso de CONECTORES LOGICOS &

```python

#los registros cuyo review_scores_cleanliness sea menor a 3 y review_scores_rating sea mayor a 3
#para esto solo necesitamos indicar al operacion que se quiere hacer
filtro8=df[(df["review_scores_cleanliness"]<3) & (df["review_scores_rating"]>3)]
filtro8
```
